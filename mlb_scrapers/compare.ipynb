```python
import pandas as pd
import numpy as np

true_odds_before = pd.read_csv('2024-06-30_12-06-19.csv')
true_odds_after = pd.read_csv('2024-06-30_14-52-19.csv')
```

```python
merged_df = pd.merge(true_odds_before, true_odds_after, on = ['players', 'bet_type'], how = 'inner')
```

```python
merged_df
```

```python
merged_df['diff'] = merged_df['over_true_prob_y'] - merged_df['over_true_prob_x']
```

```python
merged_df[(np.abs(merged_df['diff']) > 0.05) & (merged_df['over_lines_x'] == merged_df['over_lines_y'])]
```

```python
print(len(merged_df[merged_df['diff'] == 0]) / len(merged_df))

print(np.mean(np.abs(merged_df[merged_df['diff'] != 0]['diff'])))
```

```python
moved = merged_df[merged_df['diff'] != 0]
moved["abs_diff"] = np.abs(moved['diff'])
```

```python
moved.groupby("bet_type")['abs_diff'].agg(['median', 'count', 'std', 'mean'])
```
